```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./Data/humans_only_absent.csv')
data.head()
```

```python
sns.distplot(data['DLIndex'])
```

```python
import FRA
```

```python
regions = [FRA.code2Vector(FRA.lettercode2Strategy(r, 8), 8) for r in FRA.regionsCoded]
```

```python
cols = ['a' + str(i) + str(j) for i in range(1, 9) for j in range(1, 9)]
data['Region'] = data.apply(lambda x: list(x[cols]), axis=1)
data['Dists'] = data['Region'].apply(lambda x: FRA.minDist2Focal(x, regions))
data.head()
```

```python
df_high_dl = pd.DataFrame(data[data['DLIndex'] > 0.98])
fig, ax = plt.subplots(figsize=(2,2))
sns.distplot(df_high_dl['Dists'], kde=False, ax=ax)
ax.set_xlabel('Number of misplaced tiles')
ax.set_ylabel('Frequency')
ax.set_title('Selected regions where\n DLindex > 0.98')
```

```python
tolerance = 0
non_shaky = df_high_dl[df_high_dl['Dists'] <= tolerance]['Dists'].count()
shaky = df_high_dl[df_high_dl['Dists'] > tolerance]['Dists'].count()
print("Shaky:", shaky, "Non shaky:", non_shaky)
print("Shaky hand percentage:", shaky/(shaky + non_shaky))
print("Non shaky hand percentage:", non_shaky/(shaky + non_shaky))
```

```python
tolerance = 3
non_shaky = df_high_dl[df_high_dl['Dists'] <= tolerance]['Dists'].count()
shaky = df_high_dl[df_high_dl['Dists'] > tolerance]['Dists'].count()
print("Shaky:", shaky, "Non shaky:", non_shaky)
print("Shaky hand percentage:", shaky/(shaky + non_shaky))
print("Non shaky hand percentage:", non_shaky/(shaky + non_shaky))
```

```python
cols = ['Dyad', 'Player', 'Round', 'Score', 'Region'] + cols 
data = data[cols]
data.head()
```

```python
tolerancia = 0
data['Category'] = data['Region'].apply(lambda x: FRA.classify_region(x, regions, tolerancia))
data['RegionGo'] = data.groupby('Player')['Category'].transform('shift', -1)
data.head()
```

```python
data.to_csv('./Data/humans_tolerance0.csv')
```
